```python
import requests
from bs4 import BeautifulSoup
import time
import numpy as np
import pandas as pd
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/146.0.0.0 Safari/537.36', "Accept-Language": "en-US,en;q=0.9",
    "Referer": "https://www.google.com/"}
```

```python
def get_review_stat(soup, class_name):
    container = soup.find('div', class_=class_name)
    if container and container.find('strong'):
        return container.find('strong').text.strip()
    return "0"
```

```python
def get_review_text(soup, num):
  containers = soup.find_all('div', class_ = 'thumbbody mt8')
  if containers and len(containers) > num:
    text_div = containers[num].find('div', class_='text')
    if text_div:
      return text_div.text.strip()
    return "N/A"
```

```python
animeler = []
```

```python
print(len(animeler))
```

```python
for i in range(0, 10100, 50):
  url = f"https://myanimelist.net/topanime.php?limit={i}"
  try:
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')
    anime_list = soup.find_all('tr', class_='ranking-list')
    for anime in anime_list:
      j = i + anime_list.index(anime)
      print(f'Now Im on {j}')
      title = anime.find('h3', class_='fl-l fs14 fw-b anime_ranking_h3')
      title = title.text if title else 'N/A'
      link_tag = anime.find('a', class_='hoverinfo_trigger')
      if link_tag:
        anime_url = link_tag.get('href')
        page = requests.get(anime_url, headers = headers)
        page_soup = BeautifulSoup(page.text, 'html.parser')
        score = page_soup.find('div', class_="score-label")
        score = score.text if score else 'N/A'
        people_scored = page_soup.find('span', itemprop="ratingCount")
        people_scored = people_scored.text if people_scored else 'N/A'
        ranked = page_soup.find('span', class_="numbers ranked")
        ranked = ranked.text if ranked else 'N/A'
        popularity = page_soup.find('span', class_="numbers popularity")
        popularity = popularity.text if popularity else 'N/A'
        members = page_soup.find('span', class_="numbers members")
        members = members.text if members else 'N/A'
        season = page_soup.find('span', class_="information season")
        season = season.text if season else 'N/A'
        title_type = page_soup.find('span', class_="information type")
        title_type = title_type.text if title_type else 'N/A'
        studio = page_soup.find('span', class_="information studio author")
        studio = studio.text if studio else 'N/A'
        description = page_soup.find('p', itemprop="description")
        description = description.text if description else 'N/A'
        divs = page_soup.find_all('div', class_ = 'spaceit_pad')
        anime_data = {}
        for div in divs:
          div = div.text
          if ":" in div:
            key, value = div.split(":", 1)
            anime_data[key.strip()] = value.strip()
        rec = get_review_text(page_soup, 0)
        mix = get_review_text(page_soup, 1)
        not_rec = get_review_text(page_soup, 2)
        recn = get_review_stat(page_soup, 'recommended')
        mixn = get_review_stat(page_soup, 'mixed-feelings')
        not_recn = get_review_stat(page_soup, 'not-recommended')
        time.sleep(np.random.randint(1, 3))
        bir_anime = {
          'title': title,
          'score': score,
          'people_scored': people_scored,
          'ranked': ranked,
          'popularity': popularity,
          'members': members,
          'season': season,
          'title_type': title_type,
          'studio': studio,
          'description': description,
          'anime_data': anime_data,
          'recommended': recn,
          'mixed_feelings': mixn,
          'not_recommended': not_recn,
          'rec': rec,
          'mix': mix,
          'not_rec': not_rec
        }
        animeler.append(bir_anime)
    time.sleep(1)
  except Exception as e:
    print(f"Error at limit {i}: {e}")

  df = pd.DataFrame(animeler)
  df.to_csv('anime_dataset3.csv', index = False)
```

```python
extras = extras.str.replace("'Genre'", "'Genres'")
extras = extras.str.replace("'Theme'", "'Themes'")
extras = extras.str.replace("'Demographic'", "'Demographics'")
```

```python
import ast
```

```python
def extract_thingy(extra, key):
  try:
    d = ast.literal_eval(extra)
    if key in d.keys():
        return d[key]
  except (SyntaxError, ValueError, TypeError):
    return 'N/A'
```

```python
df['english_title'] = extras.apply(extract_thingy, key = 'English')
df['show_type'] = extras.apply(extract_thingy, key = 'Type')
df['status'] = extras.apply(extract_thingy, key = 'Status')
df['aired'] = extras.apply(extract_thingy, key = 'Aired')
df['season'] = extras.apply(extract_thingy, key = 'Premiered')
df['broadcast'] = extras.apply(extract_thingy, key = 'Broadcast')
df['producers'] = extras.apply(extract_thingy, key = 'Producers')
df['licensors'] = extras.apply(extract_thingy, key = 'Licensors')
df['source'] = extras.apply(extract_thingy, key = 'Source')
df['genres'] = extras.apply(extract_thingy, key = 'Genres')
df['duration'] = extras.apply(extract_thingy, key = 'Duration')
df['episodes'] = extras.apply(extract_thingy, key = 'Episodes')
df['rating'] = extras.apply(extract_thingy, key = 'Rating')
df['themes'] = extras.apply(extract_thingy, key = 'Themes')
df['demographics'] = extras.apply(extract_thingy, key = 'Demographics')
df['favorites'] = extras.apply(extract_thingy, key = 'Favorites')
```

```python
df = df.drop(columns = ['anime_data'])
```
